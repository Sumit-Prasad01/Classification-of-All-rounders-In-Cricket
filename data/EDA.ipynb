```python
df.head()
```

```python
df_new = pd.DataFrame(df)

batting_threshold = df_new['Batting Ave'].mean()
bowling_threshold = df_new['Bowling Ave'].mean()
```

```python
def classify_player(row):
    if row['Batting Ave'] >= batting_threshold and row['Bowling Ave'] <= bowling_threshold:
        return 'Balanced All-Rounder'
    elif row['Batting Ave'] >= batting_threshold:
        return 'Batting All-Rounder'
    elif row['Bowling Ave'] <= bowling_threshold:
        return 'Bowling All-Rounder'
    else:
        return 'Others'

df_new['Category'] = df_new.apply(classify_player, axis=1)

print(df_new)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_new, x='Batting Ave', y='Bowling Ave', hue='Category', style='Category', s=100)
plt.title('All-Rounder Categorization (Batting vs Bowling Avg)')
plt.xlabel('Batting Average')
plt.ylabel('Bowling Average')
plt.grid(True)
plt.gca().invert_yaxis()  # Lower bowling avg is better
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df_new, x='Category', palette='Set2')
plt.title('Count of All-Rounder Categories')
plt.xticks(rotation=15)
plt.ylabel("Number of Players")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df_new['Category'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Distribution of Player Categories')
plt.ylabel('')
plt.show()
```
